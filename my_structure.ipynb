```python
#4 через словник (dict)

cars = [
    {"марка": "Toyota Corolla", "рік": 2018, "ціна": 35000},
    {"марка": "BMW X5", "рік": 2020, "ціна": 65000},
    {"марка": "Renault Logan", "рік": 2016, "ціна": 27000},
    {"марка": "Mazda 3", "рік": 2019, "ціна": 39000},
]

print("Наявність на складі:")
for car in cars:
    print(f"Марка: {car['марка']}, Рік: {car['рік']}, Ціна: {car['ціна']} грн")

print("\nМашини з ціною менше 40000 грн:")
for car in cars:
    if car["ціна"] < 40000:
        print(f"{car['марка']} ({car['рік']})")
```

```python
#4 через клас (class)

class CarStock:
    авто = [
        {"марка": "Toyota Corolla", "рік": 2018, "ціна": 35000},
        {"марка": "BMW X5", "рік": 2020, "ціна": 65000},
        {"марка": "Renault Logan", "рік": 2016, "ціна": 27000},
        {"марка": "Mazda 3", "рік": 2019, "ціна": 39000},
    ]

print("Наявність на складі:")
for car in CarStock.авто:
    print(f"Марка: {car['марка']}, Рік: {car['рік']}, Ціна: {car['ціна']} грн")

print("\nМашини з ціною менше 40000 грн:")
for car in CarStock.авто:
    if car["ціна"] < 40000:
        print(f"{car['марка']} ({car['рік']})")
```
